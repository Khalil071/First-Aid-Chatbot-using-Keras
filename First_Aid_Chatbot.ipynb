```python
voc = []
tags = []
doc_x = []
doc_y = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        tokens = nltk.word_tokenize(pattern)
        voc.extend(tokens)
        doc_x.append(pattern)
        doc_y.append(intent['tag'])

    if intent['tag'] not in tags:
        tags.append(intent['tag'])
```

```python
voc = [word for word in voc if word not in string.punctuation]

lemmatizer = WordNetLemmatizer()
voc = [lemmatizer.lemmatize(word.lower() ) for word in voc]

stop_words = stopwords.words('english')
voc = [word for word in voc if word not in stop_words]

voc = set(voc)
voc = sorted(voc)
voc
```

```python
tags = sorted(set(tags))
tags
```

```python
import random
import numpy as np

x = []
y = []

for idx, doc in enumerate(doc_x):
    bow = []
    doc = lemmatizer.lemmatize(doc.lower() )
    
#It converts doc into a binary vector, where 1 indicates the word is present, and 0 means it is absent.
    for word in voc:
        bow.append(1) if word in doc else bow.append(0)

    x.append(bow)
    output_row = [0]*len(tags)
    cl = doc_y[idx]
    cl_idx = tags.index(cl)
    output_row[cl_idx] = 1

    y.append(output_row)
```

```python
from keras import Sequential
from keras.layers import Input, Dense, Dropout

input_shape = (len(x[0]), )
output_shape = len(y[0])

model = Sequential()
model.add(Input(shape=input_shape))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(output_shape, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=200)
```

```python
import nltk

def clean_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [lemmarizer.lemmatize(word) for word in tokens]
    
    stop_words = stopwords.words('english')
    tokens = [word for word in tokens if word not in stop_words]
    return tokens
```

```python
def words_bag (text, voc):
    tokens = clean_text(text)
    bow = [0] * len(voc)
    for w in tokens:
        for idx, word in enumerate(voc):
            if word == w:
                bow[idx] = 1

    return bow
```

```python
def pred_tag(text, model, voc, tags):
    bow = words_bag(text, voc)
    predictions = model.predict([bow], verbose=0)[0]

    max_idx = np.argmax(predictions)
    return tags[max_idx]
```
